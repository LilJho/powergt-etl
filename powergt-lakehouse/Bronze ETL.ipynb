#INIT

```python
from pyspark.sql.functions import current_timestamp
```

#Read The Data from the volume

```python
loans_df = (
    spark.read.option("header", "true")
    .option("inferSchema", "true")
    .csv("/Volumes/workspace/bronze/source_systems/pps_pawshop_src/loans.csv")
)

loancharges_df = (
    spark.read.option("header", "true")
    .option("inferSchema", "true")
    .csv("/Volumes/workspace/bronze/source_systems/pps_pawshop_src/loancharges.csv")
)

loanschemecharges_df = (
    spark.read.option("header", "true")
    .option("inferSchema", "true")
    .option("quote", '"')
    .option("escape", '"')
    .option("multiLine", "true")
    .option("mode", "PERMISSIVE")
    .csv("/Volumes/workspace/bronze/source_systems/pps_pawshop_src/loanschemecharges.csv")
)

loanchargetypes_df = (
    spark.read.option("header", "true")
    .option("inferSchema", "true")
    .option("quote", '"')
    .option("escape", '"')
    .option("multiLine", "true")
    .option("mode", "PERMISSIVE")
    .csv("/Volumes/workspace/bronze/source_systems/pps_pawshop_src/loanchargetypes.csv")
)

loandiscounts_df = (
    spark.read.option("header", "true")
    .option("inferSchema", "true")
    .csv("/Volumes/workspace/bronze/source_systems/pps_pawshop_src/loandiscounts.csv")
)

loanschemediscounts_df = (
    spark.read.option("header", "true")
    .option("inferSchema", "true")
    .option("quote", '"')
    .option("escape", '"')
    .option("multiLine", "true")
    .option("mode", "PERMISSIVE")
    .csv("/Volumes/workspace/bronze/source_systems/pps_pawshop_src/loanschemediscounts.csv")
)

loandiscounttypes_df = (
    spark.read.option("header", "true")
    .option("inferSchema", "true")
    .option("quote", '"')
    .option("escape", '"')
    .option("multiLine", "true")
    .option("mode", "PERMISSIVE")
    .csv("/Volumes/workspace/bronze/source_systems/pps_pawshop_src/loandiscounttypes.csv")
)

loanschemes_df = (
    spark.read
    .option("multiLine", "true")
    .json("/Volumes/workspace/bronze/source_systems/pps_pawshop_src/loanschemes.json")
)

loanstatuslogs_df = (
    spark.read.option("header", "true")
    .option("inferSchema", "true")
    .option("quote", '"')
    .option("escape", '"')
    .option("multiLine", "true")
    .option("mode", "PERMISSIVE")
    .csv("/Volumes/workspace/bronze/source_systems/pps_pawshop_src/loanstatuslogs.csv")
)
```

#Add EtlAddOn

```python
loans_df = loans_df.withColumn("EtlAddOn", current_timestamp())
loancharges_df = loancharges_df.withColumn("EtlAddOn", current_timestamp())
loandiscounts_df = loandiscounts_df.withColumn("EtlAddOn", current_timestamp())
loanschemecharges_df = loanschemecharges_df.withColumn("EtlAddOn", current_timestamp())
loanschemediscounts_df = loanschemediscounts_df.withColumn("EtlAddOn", current_timestamp())
loanchargetypes_df = loanchargetypes_df.withColumn("EtlAddOn", current_timestamp())
loandiscounttypes_df = loandiscounttypes_df.withColumn("EtlAddOn", current_timestamp())
loanschemes_df = loanschemes_df.withColumn("EtlAddOn", current_timestamp())
loanstatuslogs_df = loanstatuslogs_df.withColumn("EtlAddOn", current_timestamp())
```

#Load the tables in the Bronze

```python
loans_df.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("bronze.bronze_loans")
loancharges_df.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("bronze.bronze_loancharges")
loandiscounts_df.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("bronze.bronze_loandiscounts")
loanschemecharges_df.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("bronze.bronze_loanschemecharges")
loanschemediscounts_df.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("bronze.bronze_loanschemediscounts")
loanchargetypes_df.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("bronze.bronze_loanchargetypes")
loandiscounttypes_df.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("bronze.bronze_loandiscounttypes")
loanschemes_df.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("bronze.bronze_loanschemes")
loanstatuslogs_df.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable("bronze.bronze_loanstatuslogs")
```
